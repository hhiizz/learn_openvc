```python
import cv2 ,numpy
from google.colab.patches import cv2_imshow
image = cv2.imread("./123.jpg")
cv2.line(image,(50,150),(50,200),(255,0,0),2)
cv2.rectangle(image,(120,200),(180,213),(0,55,255),12)
cv2.rectangle(image,(124,200),(180,13),(12,55,55),-1)
cv2.circle(image,(500,300),40,(255,255,0),-1)
pts = numpy.array([[150,50],[250,100],[150,250],[50,100]])
cv2.polylines(image,[pts],True,(0,255,255),2)
cv2.putText(image,"llll",(130,131),cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(255,255,125),2)
cv2_imshow(image)
```

```python
fier = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
for i in range(1,4):
  image = cv2.imread('./'+str(i)+'.jpg')
  faces = fier.detectMultiScale(image,scaleFactor=3.1,minNeighbors=5,minSize=(30,30))
  height = image.shape[0]
  width = image.shape[1]

  cv2.putText(image,"find face"+str(len(faces)),(10,width-5),cv2.FONT_HERSHEY_COMPLEX,0.5,2)

  for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(128,255,0),6)
    image1 = image[y:y+h,x:x+w]
    image2 = cv2.resize(image1,(400,400))
    cv2.imwrite("./image"+str(i)+"faces.jpg",image2)

  cv2_imshow(image)
  cv2_imshow(image2)
```
